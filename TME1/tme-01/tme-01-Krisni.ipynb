<font color="RED" size="+1">**[Q]**</font> Dans la boîte "Code" suivante, écrire la fonction <tt>calcul</tt> qui, étant donné un réel $x$ de [0,1] rend la valeur du calcul $$-x\log_2(x)-(1-x)\log_2(1-x)$$ si $x\not= 0$ et $x \not= 1$ ou la valeur $0.0$ sinon.

```python
import math

def calcul(val) :
    if val == 0 or val == 1 :
        return 0
    if val < 0 or val > 1 :
        print("Erreur : val doit etre inclus dans l'intervalle ]0,1[")
        return
    return -val * math.log2(val) - (1 - val) * math.log2(1 - val)
```

```python
calcul(1/3)
```

Nous allons maintenant écrire les fonctions permettant de calculer des mesures ''basiques'' sur des données unidimensionnelles. Pour cela, nous allons travailler avec la structure <tt>array</tt> de numpy. Pour convertir une colonne de DataFrame en '<tt>array</tt>, voici commment faire:

### Moyenne

<font color="RED" size="+1">**[Q]**</font> **La moyenne**


Ecrire la fonction moyenne(valeurs) qui calcule la moyenne des éléments d'un tableau (en utilisant la boucle ''for'', sans utiliser les fonctions déjà implémentées dans numpy)

```python
def moyenne(tab) :
    moy = 0.
    longueur = tab.size
    for i in range(longueur) :
        moy += tab[i]
    return moy / longueur
```

Tester la fonction moyenne et comparer vos résultats avec l'implémentation de base de python donnée ci-dessous:

```python
print("La moyenne (MedQ) est avec ma fonction : %f dollars" % moyenne(prices_pd["MedQ"]))
print("La moyenne (MedQ) est avec mean        : %f dollars" % prices_pd["MedQ"].mean())
```

<font color="RED" size="+1">**[Q]**</font> **Moyenne sur les états**

Calculez la moyenne des prix qualités medium et high dans tous les états -- la liste des états est obtenues ainsi states=np.unique(prices_pd["State"].values)

Pour cela, vous pouvez (devez) le faire de deux manières:
* Faire une boucle sur chacun des états
* Utiliser la fonction groupby comme expliqué ici : http://pandas.pydata.org/pandas-docs/stable/groupby.html
 et ici : https://www.kaggle.com/crawford/python-groupby-tutorial


```python
def moyenne_state_medQ(tab) :
    states = np.unique(tab["State"].values)
    nbStates = states.size
    
    print("Prix moyen par état pour la qualité Medium (avec ma fonction) :\n")
    
    for i in range(nbStates) :
        state = states[i]
        val_moy = 0.0
        nb_val = 0
        
        for j in range(tab.shape[0]) :
            if (tab["State"][j] == state) :
                val_moy += tab["MedQ"][j]
                nb_val += 1
                
        print("\t%s : %f" % (state, (val_moy / nb_val)))
        
def moyenne_state_highQ(tab) :
    states = np.unique(tab["State"].values)
    nbStates = states.size
    
    print("Prix moyen par état pour la qualité High (avec ma fonction) :\n")
    
    for i in range(nbStates) :
        state = states[i]
        val_moy = 0.0
        nb_val = 0
        
        for j in range(tab.shape[0]) :
            if (tab["State"][j] == state) :
                val_moy += tab["HighQ"][j]
                nb_val += 1
                
        print("\t%s : %f" % (state, (val_moy / nb_val)))

        
def moyenne_state_medQ_groupby(tab) :
    return tab.groupby("State")["MedQ"].mean()
    
def moyenne_state_highQ_groupby(tab) :
    return tab.groupby("State")["HighQ"].mean()
```

```python
print("Calcul des moyennes avec ma fonction :\n")
moyenne_state_medQ(prices_pd)
print()
moyenne_state_highQ(prices_pd)
```

```python
print("Calcul des moyennes avec groupby :\n")
print("Prix moyen par état pour la qualité Medium (avec groupby) :\n")
print(moyenne_state_medQ_groupby(prices_pd))
print("Prix moyen par état pour la qualité High (avec groupby) :\n")
print(moyenne_state_highQ_groupby(prices_pd))
```

<font color="RED" size="+1">**[R]**</font> **Réponse à la question**

Le paramètre inplace permet d'écrire sur les données qu'on utilise. Si **inplace=False**, on crée une copie de l'instance en cours d'utilisation. Il faut donc créer une nouvelle variable afin de conserver en mémoire la modification sinon la copie est perdue.

Si on print directement "prices_pd.sort_values(by=['State', 'date'], inplace=False)", on remarquera que les données ne seront pas organisées.

Dans le cas contraire, lorsque **inplace=True**, on modifie directement les données. Il n'y a donc pas besoin de créer une nouvelle variable afin de conserver la modification.

Si on print directement "prices_pd.sort_values(by=['State', 'date'], inplace=True)", on remarquera que les données ont bien été organisées.

<font color="RED" size="+1">**[Q]**</font> **Changements des résultats**


Recalculer la moyenne des prix pour la qualité <tt>Low</tt>. Qu'en est-il maintenant ? 

```python
import math

def moyenne_state_lowQ(tab) :
    states = np.unique(tab["State"].values)
    nbStates = states.size
    list_of_low_prices = []
    
    for i in range(nbStates) :
        state = states[i]
        val_moy = 0.0
        nb_val = 0
        
        for j in range(tab.shape[0]) :
            if (tab["State"][j] == state) :
                val = tab["LowQ"][j]
                if (math.isnan(val) == False) :
                    val_moy += val
                    nb_val += 1
                
        list_of_low_prices.append([state, (val_moy / nb_val)])
        
    return pd.DataFrame(list_of_low_prices, columns=['State', 'LowQ'])
        
def moyenne_state_lowQ_groupby(tab) :
    
    return tab[["State","LowQ"]].groupby(["State"]).mean()
```

```python
print("Prix moyen par état pour la qualité Low (avec ma fonction) :")
print(moyenne_state_lowQ(prices_pd))
print()
print("Prix moyen par état pour la qualité Low (avec groupby) :")
print(moyenne_state_lowQ_groupby(prices_pd))
```

La fonction **fillna** permet de remplir les cases NaN, soit non-assignées, du datagramme. On décide de remplir à 0 les cases NaN.

<font color="RED" size="+1">**[Q]**</font> **Calcul d'un effectif**

Prendre un intervalle de discrétisation de taille 20, et calculer l'effectif (sous forme d'un vecteur) du nombre d'états par ''bins''. Dessiner l'histogramme correspondant

```python
# Calcul des effectifs
val_min = (min(prix_moyens))
val_max = (max(prix_moyens))
bins = 20

interv = ((val_max - val_min) / bins)
val = val_min

list_interv = [] # Liste contenant les différents intervalles
list_nb = [] # Liste contenant l'effectif par intervalle

# Les effectifs et intervalles sont parallèles entre les deux tableaux (l'effectif de l'intervalle list_interv[0]
# list_nb[0] )

while (val <= val_max) :
    interv_string = str(val) + "-" + str(val + interv)
    list_interv.append(interv_string)
    
    nb = 0
    
    for i in range(len(prix_moyens)) :
        price = prix_moyens[i]
        
        if (price >= val and price < (val + interv)) :
            nb += 1
    
    list_nb.append(nb)
    
    val += interv
    
# Histogramme
import matplotlib.pyplot as plt
plt.rcdefaults()
import numpy as np
import matplotlib.pyplot as plt

plt.rcdefaults()
fig, ax = plt.subplots()

# Mise en place des données
intervalles = list_interv
sy_pos = np.arange(len(intervalles))
nb_interv = list_nb

# Organisation des données
ax.barh(sy_pos, nb_interv, color='red')
ax.set_yticks(sy_pos)
ax.set_yticklabels(intervalles)
ax.invert_yaxis()  # labels read top-to-bottom
ax.set_xlabel('Effectif par plages de valeurs (20 bins)')
ax.set_title('Répartition des Etats en fonction du prix moyen pour la qualité Low')

# Affichage
plt.show()
```
